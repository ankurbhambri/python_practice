```python
'''

python program --Communicate using library (sqlite3 -- part of standard installation, not required to install)--> SQlite Database
python program --Communicate using library (spymysql)--> MySQL Database
python program --Communicate using library (cx-oracle)--> Oracle Database

sqlite database is as part of installtion, not required to install

my_url = "https://www.jafsoft.com/searchengines/log_sample.html"

1. Read from website url 
2. and extract IP and Pics using regular expression
3. and send to SQL-DB -(sqlite)

Step-1 : Create databse and table
Step-2 : Extract ip and pics
Step-3 : send to db

'''

import sqlite3

con = sqlite3.connect("my_database.sqlite3") # If my_database.sqlite3 file present use it else create one file

# Other SLQ-DB like mysql,oracle etc, changes are

# import pymysql
# con = pymysql.connect(user="",password="",host="",port="",db="")

# obtain cursor -> This helps to send query and retreive result

cur = con.cursor()

my_query = "CREATE TABLE IF NOT EXISTS MY_WEBSITE_DATA (IP VARCHAR(100),PICS VARCHAR(100))"
cur.execute(my_query)

my_url = "https://www.jafsoft.com/searchengines/log_sample.html"

# to read data from website, make use url lib -> we will get page source

import re # regular expression
import urllib.request as u

F = u.urlopen(my_url)

#complete_website_data = F.read()
#print(complete_website_data)

# 123.123.123.123 - - [26/Apr/2000:00:23:48 -0400] "GET /pics/wpaper.gif HTTP/1.0" 200 6248 "<A HREF="http://www.jafsoft.com/asctortf/">http://www.jafsoft.com/asctortf/</A>" "Mozilla

# inside re, there are functions
# findall --> based pattern, we can find all occurance
# split --> based pattern we can split
# match --> match till \n
for my_line in F:
    #m = re.match("which pattern?","on which string?")
    #.* (0 or more any char)
    #\s --> space
    #\S->non-space
    #\s* --> 0 or moretimes space
    #\s+ --> 1 or moretimes space
    #\s? --> 0 or 1 time space
    #\w --> [a-zA-Z0-9_]
    # \w+ --> one or more char from [a-zA-Z0-9_]
    my_line=my_line.decode() # to str
    #print(type(my_line))
    
    # 1   2    3       .     123            .   123.123 - - [26/Apr/2000:00:23:48 -0400] "GET /pics/wpaper.gif HTTP/1.0" 200 6248 "<A HREF="http://www.jafsoft.com/asctortf/">http://www.jafsoft.com/asctortf/</A>" "Mozilla/4.05 (Macintosh; I; PPC)"
    # \d  \d   \d      \. \d  \d   \d      \.
    # \d{3}\.\d{3}\.
    # 10.125.2.10
    # \d{1,3}\.\d{1,3}\.
    
    m = re.match("(\d\d\d\.\d{3}\.\d{1,3}\.[0-9]{3}).*GET\s+(/pics/(\w+\.(jpg|gif|png)))?.*",my_line)
    # Whicherver the line is not having above pattern that time m value is None
    # if m value is not None then capture ip and pics
    
    if m !=None:
        ip = m.group(1)
        pics = m.group(3)
        if pics == None:
            pics = "No Image"
        print(ip,pics)
        my_query = f"INSERT INTO MY_WEBSITE_DATA VALUES('{ip}','{pics}')" # f--> format
        cur.execute(my_query)
con.commit()
cur.execute("SELECT * FROM MY_WEBSITE_DATA")
db_result = cur.fetchall()
print("db_result : ",db_result)
    

```
